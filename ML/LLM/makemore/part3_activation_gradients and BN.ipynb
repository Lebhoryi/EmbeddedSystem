```python
import torch
```

```python
x = torch.randn(1000, 10)
w = torch.randn(10, 200) / 10**0.5

print(x.mean(), x.std())
print(w.mean(), w.std())

h = x @ w
print(h.mean(), h.std())

```

```python
import torch, numpy as np
import matplotlib.pyplot as plt

# 若 h 未计算则计算
try:
    h
except NameError:
    h = x @ w

def to_np(t):
    return t.detach().cpu().numpy().ravel() if isinstance(t, torch.Tensor) else np.asarray(t).ravel()

x_np, w_np, h_np = map(to_np, (x, w, h))

fig, axes = plt.subplots(1, 3, figsize=(12, 3))
for arr, ax, name in [(x_np, axes[0], 'x'), (w_np, axes[1], 'w'), (h_np, axes[2], 'h')]:
    ax.hist(arr, bins=100, density=True, color='#4C78A8', alpha=0.85)
    ax.set_title(f'{name}: μ={arr.mean():.3f}, σ={arr.std():.3f}')
    ax.grid(True, ls='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
